```python
import numpy as np
import matplotlib.pyplot as plt

point1 = np.array([1, 2, 3])
point2 = np.array([4, 8, 3])
point3 = np.array([0, 6, 6])
point4 = np.array([8, 7, 7])
print(np.linalg.norm(point1-point2))
print(np.linalg.norm(point2-point3) ** 2)
print(np.linalg.norm(point3-point4, ord=np.inf))
print(np.linalg.norm(point4-point1, ord=1))
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(point1[0], point1[1], point1[2])
ax.scatter(point2[0], point2[1], point3[2])
ax.scatter(point3[0], point3[1], point3[2])
ax.scatter(point4[0], point4[1], point4[2])
```
