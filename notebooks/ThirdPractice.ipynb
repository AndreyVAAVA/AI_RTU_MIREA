## Задание 2.3.1

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import seaborn as sns
import pandas as pd


iris = sns.load_dataset('iris')
#iris = pd.read_csv('C:\\Users\\ballg\\PycharmProjects\\ai_mirea\\iris.csv')
X_train, X_test, y_train, y_test = train_test_split(
    iris.iloc[:, :-1],
    iris.iloc[:, -1],
    test_size = 0.15
)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
def train_model(k, X_train, y_train, X_test):
    model = KNeighborsClassifier(n_neighbors = k)
    model.fit(X_train, y_train)
    predicted = model.predict(X_test)
    return predicted


def draw_graph(y_test, X_test, y_pred):
    plt.figure(figsize = (10, 7))
    sns.scatterplot(x = 'petal_width', y = 'petal_length', data = iris, hue = 'species', s = 70)
    plt.xlabel('Длина лепестка, см')
    plt.ylabel('Ширина лепестка, см')
    plt.legend(loc = 2)
    plt.grid()

    for i in range(len(y_test)):
        if np.array(y_test)[i] != y_pred[i]:
            plt.scatter(X_test.iloc[i, 3], X_test.iloc[i, 2], color ='red', s = 150)
```

```python
pred = train_model(1, X_train, y_train, X_test)
draw_graph(y_test, X_test, pred)
```

```python
pred = train_model(5, X_train, y_train, X_test)
draw_graph(y_test, X_test, pred)
```

```python
pred = train_model(10, X_train, y_train, X_test)
draw_graph(y_test, X_test, pred)
```
